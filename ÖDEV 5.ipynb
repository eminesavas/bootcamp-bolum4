```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
import seaborn as sns
import scipy.stats as stats
from scipy.stats.mstats import winsorize
import warnings

warnings.filterwarnings('ignore')
sns.set(style="whitegrid")

plt.rcParams['figure.dpi'] = 100
plt.rcParams['figure.figsize'] = (8,5.5)
```

## Doldurulmuş dosyamızla çalışmak için abd_egitim_sistemi'ne atamamızı yapalım

```python
abd_egitim_sistemi = pd.read_csv("abd_egitim_sistemi_temiz.csv")
```

## Tek değişkenli analiz yaparak verileri yorumlayalım

```python
abd_egitim_sistemi['WİNSORİZE_TOTAL_REVENUE'] = winsorize(abd_egitim_sistemi['TOTAL_REVENUE'],(0,0.10))
abd_egitim_sistemi['WİNSORİZE_TOTAL_EXPENDITURE'] = winsorize(abd_egitim_sistemi['TOTAL_EXPENDITURE'],(0,0.10))
abd_egitim_sistemi['WİNSORİZE_INSTRUCTION_EXPENDITURE'] = winsorize(abd_egitim_sistemi['INSTRUCTION_EXPENDITURE'],(0,0.10))
```

```python


plt.figure(figsize=(18,15))
baslik_font = {'family':'arial','color':'darkred','weight':'bold','size':13}
eksen_font = {'family':'arial','color':'darkblue','weight':'bold','size':10}

degiskenler = ['TOTAL_REVENUE','TOTAL_EXPENDITURE','INSTRUCTION_EXPENDITURE']
winsorize_degiskenler = ['WİNSORİZE_TOTAL_REVENUE','WİNSORİZE_TOTAL_EXPENDITURE','WİNSORİZE_INSTRUCTION_EXPENDITURE']

for i in range(3):
    plt.subplot(3,3,i+1)
    plt.hist(abd_egitim_sistemi[degiskenler[i]])
    plt.title(degiskenler[i],fontdict=baslik_font)
    
for i in range(3):
    plt.subplot(3,3,i+4)
    plt.hist(abd_egitim_sistemi[winsorize_degiskenler[i]])
    plt.title(winsorize_degiskenler[i],fontdict=baslik_font)
    
for i in range(3):
    plt.subplot(3, 3, i+7)
    plt.hist(np.log(abd_egitim_sistemi[degiskenler[i]]))
    plt.title(degiskenler[i] + ' (log dönüşümlü)', fontdict=baslik_font)
             
             
             
             
```

## State kolonu CALIFORNIA olanlarla adı CALIFORNIA olan yeni bir dataframe oluşturalım

```python
CALIFORNIA=abd_egitim_sistemi[abd_egitim_sistemi.STATE=='CALIFORNIA']
CALIFORNIA
```

## Ve California eyaletinin toplam gelirleri ile toplam harcamalarının yıllara göre değişiminin doğru grafiğini çizdirelim 

```python
California=pd.DataFrame({'California':CALIFORNIA["PRIMARY_KEY"],
                          'Toplam Gelir':CALIFORNIA["TOTAL_REVENUE"],
                          'Toplam Harcama':CALIFORNIA["TOTAL_EXPENDITURE"],
                           'Yıl':CALIFORNIA["YEAR"]})
California
plt.figure(figsize=(15,7))
plt.subplot(1,2,1)
plt.plot(California['Yıl'],California['Toplam Gelir'],color='purple')
plt.ylabel('Toplam Gelir')
plt.xlabel('Yıl')
plt.title("California'nın Toplam Gelir Grafiği")
plt.subplot(1,2,2)
plt.plot(California['Yıl'],California['Toplam Harcama'],color='green')
plt.ylabel('Toplam Harcama')
plt.xlabel('Yıl')
plt.title("California'nın Toplam Harcama Grafiği")
plt.show()
```
