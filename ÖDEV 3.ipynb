```python
import pandas as pd
import numpy as np
import warnings

warnings.filterwarnings('ignore')
```

```python
abd_egitim_sistemi = pd.read_csv("states_all.csv")
```

```python
abd_egitim_sistemi.isnull().head()
```

## Ve her bir değişkenin eksik ( null ) değer oranını hesaplayalım

```python
abd_egitim_sistemi.isnull().sum()*100/abd_egitim_sistemi.shape[0]
```

## Verimizdeki yıl ( year ) sütununu unutup ,her bir gözlemin aynı yıl içerisinde yapıldığını farzedelim..Ve her bir değişken için eksik değerleri sütun ortalamalarıyla dolduralım.

```python
for sutun_adi in abd_egitim_sistemi.columns[3:]: 
    abd_egitim_sistemi[sutun_adi].fillna(abd_egitim_sistemi[sutun_adi].mean(),inplace=True)
```

```python
abd_egitim_sistemi.isnull().sum()*100/abd_egitim_sistemi.count()
```

### bu sonuç bize her bir boşluğun doldurulduğunu net göstermektedir..

## Şimdi zaman faktörünü dikkate alıp eksik verileri hangi yıl ise o yılın ortalama değerleriyle dolduralım,bunun için filtreleme ve groupby olmak üzere iki yöntemi de kullanabiliriz..

## Filtreleme ile..

```python
abd_egitim_sistemi = abd_egitim_sistemi.copy()

years = abd_egitim_sistemi["YEAR"].unique()

for col in abd_egitim_sistemi.columns[3:]:
    for year in years:
        
        abd_egitim_sistemi.loc[abd_egitim_sistemi["YEAR"]==year,col].fillna(
            
            abd_egitim_sistemi[abd_egitim_sistemi["YEAR"]==year]
            [col].mean(),inplace=True
        )
```

```python
abd_egitim_sistemi.isnull().sum()*100/abd_egitim_sistemi.count()
```

## Groupby ile..

```python
for sutun_adi in abd_egitim_sistemi.columns[3:]: 
    abd_egitim_sistemi[sutun_adi].fillna(abd_egitim_sistemi.groupby(by=sutun_adi).mean()['YEAR'], inplace=True)
```

```python
abd_egitim_sistemi.isnull().sum()*100/abd_egitim_sistemi.count()
```

### burdaki sonuçlardan da anlaşılacağı üzere her bir yıl için yapılan ortalamalarla eksik değerleri doldurma işlemi sağlanamadı...(bu bize bazı yıllar için o yılın tüm değerlerinin,boş oluğunu gösterebilir..)

## Bu seferde eksik değerleri enterpolasyon metoduyla dolduralım

```python
enterpolasyon = abd_egitim_sistemi.interpolate()
enterpolasyon 
```

```python
enterpolasyon.isnull().sum()*100/abd_egitim_sistemi.count()
```

### enterpolasyon metoduyla da gördük ki üç değişken hariç doldurma işlemi başarılı olmuş..

### sonuç olarak gördük ki bu veri kümesi için doldurma işlemi en iyi genel ortalamların alınmasıyla,sonra kısmen enterpolasyon metoduyla sağlanmış ve herbir yıl için alınan ortalamalarla yapılan doldurma işleminin ise sağlanamadığı net görülmüştür..
